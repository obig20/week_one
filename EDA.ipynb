```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Sample data for 7 days
data = {
    'Open': [0.121875, 0.098438, 0.088021, 0.086458, 0.081771, 184.720001],
    'High': [0.125000, 0.098958, 0.088542, 0.087500, 0.082292, 85.860001],
    'Low': [0.097917, 0.085417, 0.081250, 0.081771, 0.082292, 179.380005],
    'Close': [0.097917, 0.086458, 0.085417, 0.081771, 0.068750, 181.710007],
    'Adj Close': [0.096354, 0.086458, 0.085417, 0.081771, 0.071354, 181.710007]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Calculate the correlation matrix
correlation_matrix = df.corr()

# Print the correlation matrix
print(correlation_matrix)

# Visualize the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()

```

```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample data based on the provided summary
data = {
    'Category': ['Open', 'High', 'Low', 'Close', 'Adj Close'],
    'Values': [37.360464, 37.801009, 36.882561, 37.351096, 37.351096]
}

# Create a DataFrame
df = pd.DataFrame(data)

# Plot the pie chart
plt.figure(figsize=(8, 8))
plt.pie(df['Values'], labels=df['Category'], autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Stock Prices')
plt.show()
```
