```python
import matplotlib.pyplot as plt
import pandas as pd

# Sample data based on the provided summary
data = {
    'Date': pd.date_range(start='2024-07-24', periods=10993, freq='D'),
    'Open': pd.Series([	224.000000] * 10993),
    'High': pd.Series([224.800003] * 10993),
    'Low': pd.Series([217.130005] * 10993),
    'Close': pd.Series([218.539993] * 10993)
}

# Create a DataFrame
df = pd.DataFrame(data)

# Plot the data
plt.figure(figsize=(10, 6))
plt.plot(df['Date'], df['Open'], label='Open')
plt.plot(df['Date'], df['High'], label='High')
plt.plot(df['Date'], df['Low'], label='Low')
plt.plot(df['Date'], df['Close'], label='Close')

# Add labels and title
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Stock Prices Over Time')
plt.legend()

# Show the plot
plt.show()
```
